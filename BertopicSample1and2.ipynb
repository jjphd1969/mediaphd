```python
import pandas as pd

csv_path = '/content/drive/MyDrive/Collab/Julian/Reduced_Topics/tweets_reduced_topics_no_extra_cols.csv'
df = pd.read_csv(csv_path)

# Define the organizations to filter by
organizations_to_keep = ['OUTA1', 'OUTA2', 'CW1', 'CW2', 'LRC1', 'LRC2']

# Filter the DataFrame
df_filtered = df[df['organization'].isin(organizations_to_keep)].copy()

# Display the first few rows of the filtered DataFrame and the 'tweet_content' column
print("Filtered DataFrame Head:")
display(df_filtered[['organization', 'tweet_content']].head())
```

```python
# Install BERTopic library
!pip install bertopic
```

```python
from bertopic import BERTopic

# Ensure 'tweet_content' column only contains strings
df_filtered['tweet_content'] = df_filtered['tweet_content'].astype(str)

# Extract the tweet content for topic modeling
tweets = df_filtered['tweet_content'].to_list()

# Initialize BERTopic model
# Using 'english' for the default SentenceTransformer model
# You can specify a different model or custom embeddings if needed
model = BERTopic(language="english", calculate_probabilities=True, verbose=True)
```

Training the BERTopic model can take some time depending on the size of your dataset and the complexity of the embeddings.

```python
# Train the BERTopic model
topics, probs = model.fit_transform(tweets)
```

```python
import os

# Define the path where you want to save the model in your Google Drive
save_path = '/content/drive/MyDrive/Collab/Julian/bertopic_model'

# Create the directory if it doesn't exist
os.makedirs(save_path, exist_ok=True)

# Save the BERTopic model
# Using 'safetensors' for a more robust and secure serialization format.
# If you encounter issues, you can try 'pickle' as 'model.save(os.path.join(save_path, "my_bertopic_model"), serialization="pickle")'
model.save(os.path.join(save_path, "my_bertopic_model"), serialization="safetensors")

print(f"BERTopic model saved successfully to {save_path}")
```
